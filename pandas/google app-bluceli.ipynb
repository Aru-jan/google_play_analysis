```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取 CSV 文件
df = pd.read_csv('C:/Users/dell/Desktop/10-6-2_Category.csv') 

# 假设 CSV 文件有 'Category' 和 'Rating' 两列
Category = df['Category']
avg_ratings = df['avg_rate']

# 生成柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(Category, avg_ratings, color='orange')

# 在柱体顶部标注数值
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# 设置标题和轴标签
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Average Rating by Category')

# 显示图表
plt.xticks(rotation=45) # 如果类别名称太长，可以旋转x轴标签
plt.ylim([4.0,5])
plt.tight_layout()
plt.show()



```
